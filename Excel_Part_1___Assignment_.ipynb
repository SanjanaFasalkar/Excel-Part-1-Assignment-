-  Question 1: What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly explain with examples.
-  Answer:
Paste: This option simply pastes everything (values, formulas, formatting, etc.) from the copied cell(s) into the target cell.

Example: If you copy a cell containing =A1+B1 with bold formatting and paste it, both the formula and bold formatting will be pasted.

Paste Special: This gives you more control over what to paste. You can choose to paste only values, formulas, formats, or perform operations (like add, subtract, multiply).

Example: If you copy a cell containing =A1+B1 and use Paste Special → Values, only the calculated result (e.g., 15) is pasted, without the formula or formatting.

-  Question 2: Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes' in Excel.
-  Answer:
Freeze Panes: Locks specific rows or columns in place so they remain visible when scrolling.

Usefulness: Helps keep headers visible while scrolling large datasets.

Example: If you freeze the first row, column headings (like Name, Age, Salary) remain visible while you scroll down.

Split Panes: Divides the worksheet into two or four independent scrolling areas.

Usefulness: Allows you to view and compare different parts of a large sheet at the same time.

Example: You can split a sheet vertically and scroll down the bottom half while keeping the top half in view.

-  Question 3: Explain the difference between inserting a new row and inserting a new column in Excel. Can you insert multiple rows or columns at once?
-  Answer:
Inserting a Row: Adds a new row above the selected row, shifting existing rows downward.

Inserting a Column: Adds a new column to the left of the selected column, shifting existing columns to the right.

Multiple Rows/Columns: Yes, you can insert multiple rows or columns by selecting multiple rows/columns first, then right-click → Insert.

Example: Selecting 3 rows and clicking "Insert" will add 3 new rows above the selection.

-  Question 4: What are logical functions in Excel? Provide examples of at least two logical functions and their applications.
-  Answer:
Logical functions perform logical tests and return results like TRUE, FALSE, or specific values based on conditions.

IF Function:

Syntax: =IF(condition, value_if_true, value_if_false)

Example: =IF(A1>50,"Pass","Fail") → Returns "Pass" if A1 > 50, otherwise "Fail".

AND Function:

Syntax: =AND(condition1, condition2, …)

Example: =AND(A1>40, B1<100) → Returns TRUE only if both conditions are met.

OR Function (additional example):

Syntax: =OR(condition1, condition2, …)

Example: =OR(A1="Red", A1="Blue") → Returns TRUE if A1 is either Red or Blue.

-  Question 5: Discuss the purpose of 'XLOOKUP' and how it differs from the traditional 'VLOOKUP' function.
-  Answer:Purpose of XLOOKUP:
XLOOKUP is a modern lookup function in Excel that searches a range or array, finds a match, and returns the corresponding value. It is more powerful and flexible than VLOOKUP or HLOOKUP.

Differences between XLOOKUP and VLOOKUP:

Search Direction:

VLOOKUP only searches vertically (first column).

XLOOKUP can search both vertically and horizontally.

Column Index Issue:

VLOOKUP requires a column index number (e.g., 2, 3). If columns are rearranged, the formula may break.

XLOOKUP directly references the return column, so it is not affected by column order.

Exact Match:

VLOOKUP defaults to approximate match unless specified (FALSE).

XLOOKUP defaults to exact match, reducing errors.

Search from End:

XLOOKUP can search from the top or bottom of a range. VLOOKUP can only search from the top.

Error Handling:

XLOOKUP allows a built-in "if not found" message.
Example: =XLOOKUP("John", A2:A10, B2:B10, "Not Found").

VLOOKUP requires using IFERROR separately.

-  Question 10: You're working with a dataset that contains missing values. As a Data Scientist, explain how you'd detect and handle missing data using Excel.
-  Answer:Detecting Missing Data:

Go To Special:

Press Ctrl+G → Click Special → Select Blanks → All empty cells will be highlighted.

ISBLANK Function:

Formula: =ISBLANK(A2) → Returns TRUE if the cell is blank, otherwise FALSE.

COUNTBLANK Function:

Formula: =COUNTBLANK(A2:A10) → Returns the total number of blank cells in the range.

Handling Missing Data:

Delete Blanks: Remove rows/columns with too many blanks.

Replace Blanks: Use Find & Replace or enter placeholder values like 0 or N/A.

Fill Missing Data: Use techniques like filling with the column average/median or copying adjacent values.

 Example:
If a dataset has missing ages in column B, you can use =COUNTBLANK(B2:B10) to quickly count missing entries, then replace or fill them.
